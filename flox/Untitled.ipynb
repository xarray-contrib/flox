```python
import numpy as np
import xarray as xr

# Create example data
data = np.random.rand(3, 4, 5)  # Example data with dimensions (3, 4, 5)

# Create xarray dataset
ds = xr.Dataset(
    {
        "variable1": (["dim1", "dim2", "dim3"], data),  # Add variable with dimensions
        "variable2": (["dim1", "dim2", "dim3"], data),  # Add another variable with dimensions
    },
    coords={
        "dim1": np.arange(3),  # Define coordinates for dimension 1
        "dim2": np.arange(4),  # Define coordinates for dimension 2
        "dim3": np.arange(5),  # Define coordinates for dimension 3
    },
)
ds.groupby_bins("variable1", bins=[0.0, 0.5, 1.0])
```
